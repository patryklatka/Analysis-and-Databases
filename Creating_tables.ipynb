###In this exercise, I learned how to create tables in a database using SQL. Since access to the database is private, I have only included the code that creates the table below.###

db_Airbnb.png

```python
dbExecute(mysqlAGH, "CREATE TABLE users (
    id INTEGER PRIMARY KEY,
    email VARCHAR(255)
  )
")


dbExecute(mysqlAGH, "CREATE TABLE countries
(
  id INTEGER PRIMARY KEY,
  country_code VARCHAR(255),
  name VARCHAR(255)
)
")


dbExecute(mysqlAGH,"CREATE TABLE cities
(
  id integer PRIMARY KEY,
  name VARCHAR(255),
  country_id INTEGER,
  FOREIGN KEY (country_id) REFERENCES countries(id)
)
  ")


dbExecute(mysqlAGH, "CREATE TABLE hosts
(
  id INTEGER PRIMARY KEY,
  user_id INTEGER REFERENCES users(id)
)
")


dbExecute(mysqlAGH,  "CREATE TABLE places (
    id INTEGER PRIMARY KEY,
    address VARCHAR(255),
    host_id INTEGER,
    city_id INTEGER,
    FOREIGN KEY (host_id) REFERENCES hosts(id),
    FOREIGN KEY (city_id) REFERENCES cities(id)
  )
")


dbExecute(mysqlAGH,  "CREATE TABLE bookings (
    id INTEGER PRIMARY KEY,
    user_id INTEGER REFERENCES users(id),
    place_id INTEGER REFERENCES places(id),
    start_date DATE,
    end_date DATE,
    price_per_night FLOAT,
    num_nights INTEGER
  )
")


dbExecute(mysqlAGH, "CREATE TABLE reviews (
    id INTEGER PRIMARY KEY,
    booking_id INTEGER,
    rating TINYINT,
    review_body TEXT,
    FOREIGN KEY (booking_id) REFERENCES bookings(id)
  )
")
```
