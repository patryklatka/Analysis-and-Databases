###The purpose of the exercise was to learn the basic functions for visualization available within ggplot2 and tidyverse.###

```r
install.packages("tidyverse")
library(tidyverse)
patients <- read_tsv("patient-data-cleaned.txt")
patients
```

**Spot charts (scatterplot)**

**1.Draw a dot plot of the patient's weight as a function of BMI (in global aesthetics) and color the points based on height (in local aesthetics).**

**2.Using an additional geom, fit a smooth line defining the trend.**

**3.Does the fit from the previous graph look right? Look at the help page for geom_smooth and adjust the method to fit a straight line and not draw a confidence interval.**

```r
ggplot(data=patients, mapping=aes(x=BMI, y=Weight)) + geom_point(aes(colour=Height))


ggplot(data=patients, mapping=aes(x=BMI, y=Weight)) + geom_point(aes(colour=Height)) + geom_smooth()


ggplot(data=patients, mapping=aes(x=BMI, y=Weight)) + geom_point(aes(colour=Height)) + geom_smooth(method = "lm", se=FALSE)
```

**Box charts**

**1.Generate a box plot of the Score variable comparing smokers and non-smokers**

**1.Using colors, separate the box plots further by gender.**

**Tiles**

**1.Draw a dot plot of a patient's weight as a function of BMI and color the points based on height. Divide it into a grid of graphs (tiles) based on gender and smoking.**

**2.Generate a BMI boxplot comparing smokers and non-smokers, colored by gender, and add a separate tile by age.**

**3.Create a similar BMI boxplot, but this time group the data by gender, color by age (hint:convert color to a categorical variable) and tile by who is a smoker.**

```r
ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight, color=Height)) + facet_grid(Sex~Smokes)

ggplot(data=patients) + geom_boxplot(mapping=aes(x=BMI, y=Smokes, color=Sex)) + facet_grid(~Age)

ggplot(data=patients) + geom_boxplot(mapping=aes(x=BMI, y=Smokes, color=factor(Age))) + facet_wrap(~Sex)


```

**Scales**

**1.Draw a dot plot of the patient's weight as a function of BMI.**

**2.Starting from the previous graph, adjust the BMI axis so that only values of 20, 30, 40 are marked and the weight axis shows values from 60 to 100 with a step of 5. Add the Polish name of the variable and the unit (kg).**

**3.Draw a dot plot of the patient's weight as a function of BMI. Color it in a color scale due to height. Create a color scale with the middle point corresponding to the average height of the patient while the extremes of the scale are to be green (minimum) and red (maximum) in shades consistent with AGH colors. Take gray as the middle color**

```r
ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight))

ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight)) + scale_x_continuous(breaks=c(20,30,40), limits=c(20,40)) + scale_y_continuous(name="Waga [kg]",breaks=seq(60, 100, by = 5))

mean_Height <- mean(patients$Height)
ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight, color=Height)) + scale_colour_gradient2(low="green4",mid="grey", high="red3",midpoint=mean_Height)


```

**Motifs**

**1.Draw a dot plot of a patient's weight as a function of BMI. Color it on a color scale for height. Add to it trend lines without confidence intervals for each age group**

**2.Remove the title of the legend. Change the background colors of the legend position (key) to white and place the legend under the chart.**

**3.Add an appropriate title to the graph. Remove minor gridlines. Save as a 16 by 16 cm png file.**

```r
ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight, color=Height)) + geom_smooth(mapping=aes(x=BMI, y=Weight, group=Age), method="lm", se=FALSE)

ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight, color=Height)) + geom_smooth(mapping=aes(x=BMI, y=Weight, group=Age), method="lm", se=FALSE) + labs(colour=NULL, x=NULL, y=NULL ) + theme(legend.position = "bottom" ,legend.key = element_rect(fill = "white"))

ggplot(data=patients) + geom_point(mapping=aes(x=BMI, y=Weight, color=Height)) + geom_smooth(mapping=aes(x=BMI, y=Weight, group=Age), method="lm", se=FALSE) + labs(colour=NULL, x=NULL, y=NULL ) + theme(legend.position = "bottom" ,legend.key = element_rect(fill = "white")) + labs(title = "Waga pacjenta w funkcji BMI") + theme(panel.grid.minor = element_blank())

ggsave("Motywy_zad3.png")

```
