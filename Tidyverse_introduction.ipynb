###In this exercise, I learned the basics of the tidyverse package. ###

```r
install.packages("nycflights13")
install.packages("tidyverse")
library(nycflights13)
library(tidyverse)
glimpse(flights)


library(nycflights13)
library(tidyverse)
flights <- glimpse(flights)
airlines <- glimpse(airlines)
airports <- glimpse(airports)
planes <- glimpse(planes)
weather <- glimpse(weather)
```

**Had a delay of at least 2 hours.**

```r
flights  %>% filter(dep_delay > 120 | arr_delay > 120)
```

**Were flying to Huston (IAH or HOU)**

```r
flights %>% filter(dest %in% c('IAH','HOU'))
```

**Were from United, American, or Delta airlines**

```r
flights %>% inner_join(airlines) %>% filter(carrier %in% c('DL', 'AA', 'U'))
```

**Flew out in the summer (July, August, September)**

```r
flights %>% filter(month %in% c(7,8,9))
```

**Have had a delay of at least two hours but no departure delay.**

```r
flights %>% filter(dep_delay > 120 & (dep_time <= sched_dep_time) & ((sched_dep_time+(dep_delay%%60)+(dep_delay %/%60*100)) < 2360 ))
```

**Had at least an hour delay but made up more than 30 minutes on the flight.**

```r
flights %>% filter(dep_delay >= 60 & arr_delay <= -30)
```

**Sort the flights to find the flights with the longest departure delay.**

```r
flights %>% arrange(desc(dep_delay))
```

**Find the flights that departed earliest in the morning**

```r
flights %>% arrange(dep_time)
```

**Sort flights to find the fastest connections.**

```r
flights %>% arrange(air_time) %>% distinct(origin, dest, air_time)
```

**Did each day of 2013 have at least one flight?**

```r
flights %>% filter(year==2013) %>% distinct(month, day) %>% count()
```

**Which flights covered the greatest distance and which covered the shortest distance (top and bottom 10)**

```r
flights %>% distinct(distance, origin, dest) %>% arrange(distance) %>% head(10)
```

```r
flights %>% distinct(distance, origin, dest) %>% arrange(distance) %>% tail(10)
```
